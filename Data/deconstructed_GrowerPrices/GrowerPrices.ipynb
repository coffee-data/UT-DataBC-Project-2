```python
# Remove those empty rows
df1 = df.dropna(subset=['Country'])
# # Recheck
df1.isna().sum()
```

```python
# Drop rows with completely missing data
df1 = df1.dropna(how='all')
```

```python
# Iterate over currency and remove extraneous elements
def hasNumbers(inputString):
    return any(char.isdigit() for char in inputString)

s = df1['Currency_Type']

new_s = []
num = []
count = 0

for elt in s:
    elt = elt.replace('(',"")
    elt = elt.replace(')',"")
    elt = elt.replace('/'," ")
    elt = elt.replace(': ',"")
    
    if hasNumbers(elt):
        count += 1
        num.append(elt)
    else:
        num.append(elt.replace(elt,'nan'))
        
    new_s.append(elt)
    
df1['Currency_Type'] = new_s
df1.head()
```

```python
df1[['Country_Code','Unit','Type', 'Blank']] = df1['Currency_Type'].str.split(' ',expand=True)

df1.head(50)
```

```python
cols = list(df1)

# move the column to head of list using index, pop and insert
cols.insert(3, cols.pop(cols.index('Type')))
cols.insert(3, cols.pop(cols.index('Unit')))
cols.insert(3, cols.pop(cols.index('Currency_Type')))
cols.insert(3, cols.pop(cols.index('Country_Code')))
df2 = df1.loc[:,cols]
df2['Currency_Type'] = df2['Currency_Type'].str.extract('(\d+)')
df2['Unit'] = df2['Unit'].str.extract('(\D+)')
df2.rename(columns={'Currency_Type':'Weight'}, inplace=True)
df2.drop(columns='Blank', inplace=True)
df2.fillna(value={'Weight':1}, inplace=True)
```

```python
lst_cols = []

for col in df2.loc[:,'Jan-63':].columns:
    if 'unnamed' not in col.lower():
        lst_cols.append(col)
```

```python
# Clean up dates to remove issue of partial YY
s_dates = pd.Series(lst_cols)
s_dates_spl = s_dates.str.split("-",expand=True)

for i, elt in enumerate(s_dates_spl[1]):
    if int(elt) <= 99 and int(elt) > 18:
        s_dates_spl.at[i,1] = "19" + elt
    else:
        s_dates_spl.at[i,1] = "20" + elt
        
s_dates = s_dates_spl[0].str.cat(s_dates_spl[1], join='right', sep='-')
```

```python
#  inplace replace col names to avoid date confusion on last digits
for i, col in enumerate(df2.loc[:,'Jan-63':].columns):
    df2.rename(columns={str(col) : str(s_dates[i])}, inplace = True)
```

```python
file = "GrowerPrices_CLEANED.csv"
df = pd.read_csv(file)

df["unit_price(/kg)"]=""
df.head()

df.loc[df["Unit"]=="kg","unit_price(/kg)"]= (df["Price"]/ df["Weight"]).round(2)
df.loc[df["Unit"]=="L","unit_price(/kg)"]= (df["Price"]/ df["Weight"]/1016).round(2)
df.loc[df["Unit"]=="m","unit_price(/kg)"]= (df["Price"]/ df["Weight"]/1000).round(2)
df.loc[df["Unit"]=="lb","unit_price(/kg)"]= (df["Price"]/ df["Weight"]/0.45).round(2)
df.loc[df["Unit"]=="SQ","unit_price(/kg)"]= (df["Price"]/ df["Weight"]/45.9).round(2)

df = df[df['Price'].notna()]
df = df[df['Unit'].notna()]
df = df[(df[['Price']] != 0).all(axis=1)]

df.to_csv("Grower_Prices_VIZREADY.csv")
```
