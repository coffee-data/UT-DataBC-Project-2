```python
import pandas as pd
import datetime as dt
```

```python
df_grower_prices = pd.read_csv("Grower_Prices_VIZREADY_countrynameupdated.csv")
```

```python
df_exchange = pd.read_csv("../exchange_rates_VIZREADY.csv")
```

```python
df_grower_prices.drop(columns=['Unnamed: 0','Unnamed: 0.1','Unnamed: 0.1.1'], inplace=True)
```

```python
df_grower_prices.head()
```

```python
df_exchange.drop(columns='Unnamed: 0', inplace=True)
```

```python
df_exchange.rename(columns=({'Date':'Year'}), inplace=True)
```

```python
df_grower_prices['Datetime'] = pd.to_datetime(df_grower_prices["Date"], format='%b-%Y')
```

```python
df_grower_prices['Year'] = [d.year for d in df_grower_prices['Datetime']]
df_grower_prices['Month'] = [d.month for d in df_grower_prices['Datetime']]
```

```python
df_grower_prices.drop(columns=['Date'], inplace=True)
```

```python
df_grower_prices
```

```python
df_growers_merged = pd.merge(df_grower_prices, df_exchange, on=['Country', 'Year'])
```

```python
df_growers_merged.to_csv("growers_MERGED_VIZREADY.csv")
```
